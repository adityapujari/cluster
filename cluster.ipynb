```python
import pandas as pd
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=1000)
df = pd.DataFrame(X, y)
df = df.rename(columns={0: "f1", 1:"f2"})
df.head()
df.info()
```

```python
#Implementing clustering model and assign cluster labels to every record
from sklearn.cluster import DBSCAN
db = DBSCAN().fit(df)
labels = db.labels_
# Attaching labels back to the pandas dataframe
df['label'] = pd.DataFrame(labels)
df.head()
```

```python

import seaborn as sns
sns.scatterplot(x='f1', y='f2', data=df, hue='label')
```
